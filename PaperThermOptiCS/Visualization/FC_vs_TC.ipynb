```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.cm as cm
```

```python
df = pd.read_csv('FC_vs_TC.csv')
df.dropna(inplace=True)
df.reset_index(drop=True, inplace=True)
df.head()
```

```python
fig,ax=plt.subplots(2,1,figsize=(10,10))
ax=ax.flatten()
vars=['Size','nBlkd']
titles=['No. of reactions in the model','No. of thermodynamically blocked reactions']
for i in range(len(ax)):
  ax[i].plot(list(range(64)),df[['F'+vars[i]]],'o-',c='tab:green',label='FastCore')
  ax[i].plot(list(range(64)),df[['T'+vars[i]]],'o-',c='tab:blue',label='ThermOptiCS')
  ax[i].set_xticks(list(range(64)))
  ax[i].set_xticklabels([])
  ax[i].spines['top'].set_visible(False)
  ax[i].spines['right'].set_visible(False)
  ax[i].spines['bottom'].set_visible(False)
  ax[i].set_yticklabels(ax[i].get_yticklabels(),fontweight='bold',fontsize=20)
ax[0].set_ylabel('No. of reactions \nin the model',fontweight='bold',fontsize=20)
ax[1].set_ylabel('No. of thermodynamically \n blocked reactions',fontweight='bold',fontsize=20)
ax[-1].set_xticklabels(df['Model'],fontsize=6,rotation=45,fontweight='bold',ha='right',rotation_mode='anchor')
ax[-1].set_xlabel('Experiment ID of transcriptomics data',fontweight='bold',fontsize=15)
# ax[-2].set_xticklabels(df['Model'],fontsize=5,rotation=45,fontweight='bold',ha='right',rotation_mode='anchor')
handles, labels = ax[-1].get_legend_handles_labels()
tp = plt.legend([handles[0]], [labels[0]], loc=(0.65, -0.45),prop={'weight':'bold','size':20})
plt.legend([handles[1]], [labels[1]], loc=(0, -0.45),prop={'weight':'bold','size':20})
plt.gca().add_artist(tp)
plt.savefig('FCvsTC.pdf',bbox_inches='tight')
```

```python
fig,ax=plt.subplots(1,1,figsize=(10,5))
vars=['Size']
titles=['No. of reactions in the models']
temp=df[['T'+vars[0]]]
ax.plot(list(range(64)),df[['F'+vars[0]]],'o-',c='tab:green',label='FastCore')
ax.plot(list(range(64)),temp,'o-',c='tab:blue',label='ThermOptiCS')
ax.plot(list(range(64)),df['nComRxns'],'o-',c='tab:red',label='Common reactions')
ax.set_xticks(list(range(64)))
ax.set_xticklabels([])
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.set_yticklabels(ax.get_yticklabels(),fontweight='bold',fontsize=20)
ax.set_ylabel('No. of reactions \nin the model',fontweight='bold',fontsize=20)
ax.set_xticklabels(df['Model'],fontsize=6,rotation=45,fontweight='bold',ha='right',rotation_mode='anchor')
ax.set_xlabel('Experiment ID of transcriptomics data',fontweight='bold',fontsize=15)

temp = list(df['FnBlkd'])
for i,xtick in enumerate(ax.get_xticklabels()):
  if temp[i]==0:
    xtick.set_color('g')

plt.legend(prop={'size':15},loc=(0.45,0.1))
plt.savefig('FCvsTCSize.pdf',bbox_inches='tight')
```
